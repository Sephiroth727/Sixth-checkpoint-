```python
import pandas as pd
import numpy as np
dec={'name':['Anastasia','Dima','Katherine','James','Emily','Michel','Matthew','Laura','kevin','Jonas'],
          'score':[12.5,9,16.5,np.nan,9,20,14.5,np.nan,8,19],
          'attemps':[1,3,2,3,2,3,1,1,2,1],
          'qualify':['yes','no','yes','no','no','yes','yes','no','no','yes']}
labels=['a','b','c','d','e','f','g','h','i','j']
```

```python
data=pd.DataFrame(dec,index=labels)
data.head(3)
```

```python
data
```

```python
data2=data.dropna()
data2
```

```python
data2.filter(['name','score'])
```

```python
new_row={'name':'Shurech','score':15.5,'attemps':1,'qualify':'yes'}
data2=data2.append(new_row,ignore_index=True)
data2
```

```python
data3=data2.drop('attemps',axis=1)
data3
```

```python
end=[]
for i in data3['score']:
    if i > 10: end.append(1)
    else : end.append(0) 
data3['success']=end
data3
```
